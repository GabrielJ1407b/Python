```python
### Loops ###

# While

my_condition = 0

while my_condition < 10:
   print(my_condition)
   my_condition += 2
else: # Es opcional
    print("Mi condicion es mayor o igual que 10")

print("la ejecusion continua")

while my_condition < 20:
    my_condition += 1
    if my_condition == 15:
      print("Se detiene la ejecucion")
      break
    print(my_condition)

print("la ejecusion continua")

# For

my_list = [35, 24, 62, 52, 30, 17]

for element in my_list:
   print(element)

my_tuple = (35, 1.77, "Gabriel", "Garcia", "Gabrirel")

for element in my_tuple:
   print(element)

my_set = {"Gabriel", "Garcia", 16}

for element in my_set:
   print(element)

my_dict = {"Nombre": "Gabriel", "Apellido": "Garcia", "Edad": 35, 1: "Python"}

for element in my_dict:
   print(element)
   if element == "Edad":
     break
else:
   print("El bluc for para diccionario ha finalizado")

print ("La ejecucion sige")

for element in my_dict:
   print(element)
   if element == "Edad":
     continue
   print("Se ejecuto")
else:
   print("El bluc for para diccionario ha finalizado")
```
